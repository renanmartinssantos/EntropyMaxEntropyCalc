```python
import pandas as pd
import numpy as np
from scipy.stats import entropy as en
```

```python
def calcula_entropia_maxima(num_eventos):
    prob_uniforme = 1 / num_eventos
    entropia_maxima = -num_eventos * prob_uniforme * np.log2(prob_uniforme)
    return entropia_maxima
```

```python
data = pd.read_csv('beer_data.csv', low_memory=False)
#cleaning the null values
data = data.dropna(how='any',axis=0)
data.head()
```

```python
xData = data.groupby(['style']).size().reset_index(name='quantidade')
xData = xData[~xData['style'].isin(['--'])]
xData.tail()
```

```python
total = xData.sum()
print(total.quantidade)
```

```python
prob = []
for i in xData.quantidade:
    prob.append(i/total.quantidade)
prob
```

```python
entropiaCervejas = en(prob)
surpresaEsperada = 1 / entropiaCervejas
entropiaMaxima = calcula_entropia_maxima(int(total.quantidade))


print(f"A entropia dos estilos de cervejas é: {entropiaCervejas}")
print(f"A surpresa esperada ao retirar uma cerveja é: {surpresaEsperada}")
print(f"A entropia máxima dos estilos de cervejas é: {entropiaMaxima}")
```

# Conclusão

*Com base nos resultados apresentados sobre a entropia dos estilos de cervejas, podemos inferir importantes características desse universo cervejeiro. A entropia, medida de incerteza e desordem, revela um índice de 4.08, indicando uma quantidade moderada de variedade e diversidade entre os estilos. Essa pontuação sugere uma distribuição equilibrada de frequências entre as diferentes categorias de cervejas. A surpresa esperada ao retirar uma cerveja, correlacionada à entropia, é relativamente baixa (0.24), indicando que, em média, a escolha de uma cerveja não é altamente imprevisível. Os estilos de cervejas parecem seguir um padrão previsível com base na distribuição de entropia. Contrapondo-se a essa previsibilidade, a entropia máxima, alcançando 17.04, sugere um amplo espectro de estilos de cervejas com uma distribuição altamente diversificada. No entanto, a entropia real observada (4.08) está substancialmente aquém do máximo possível, indicando que a variedade de estilos não atinge seu potencial máximo. Em resumo, os resultados proporcionam insights valiosos sobre a diversidade e previsibilidade dos estilos de cervejas, onde a entropia moderada sugere uma variedade razoável e a surpresa esperada baixa indica certo grau de previsibilidade nas escolhas. A comparação com a entropia máxima destaca que, embora haja diversidade, ela ainda não atinge o ápice possível.*
